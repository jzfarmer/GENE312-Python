```python
# import our libraries
import numpy as np
import pandas as pd 
```

```python
# read in the csv file of rps strings
df = pd.read_csv("rps_strings_fall_2020.csv", header=None, names=["label","index","rps_str"])
```

```python
print(df)
```

```python
# since this is not all numbers, we will use a Python list (instead of a numpy array)
ALL_DATA = np.ndarray.tolist( df.values )

# list print out the whole thing to show its structure:
for row in ALL_DATA:
    print(row)
```

```python
# here is a function that counts substrings from BIGSTRINGS:

def count( substring, BIGSTRING ):
    """ return the number of times substring appears in BIGSTRING """
    result = 0
    for i in range(len(BIGSTRING)):
        current_slice = BIGSTRING[i:i+len(substring)]
        if substring == current_slice:
            result += 1
    return result

# let's test it by counting the number of "spr" strings that appear in "sprpsprpspr" (should be 3)
result = count("spr", "sprpsprpspr")
print("The spr-example result was", result)
```

```python
# 
# an example scoring function:  
#    here, counting the number of "r"'s plus the number of "s's" that appear in each rps string
#

def score( rps_string ):
    """ associate a score with each rps-string 
        this scoring function simply counts r's + s's
    """
    num_r = count("r", rps_string)
    num_s = count("s", rps_string)
    result_score = num_r + num_s
    return result_score


# let's test it out!
result_score = score( 'rpsrpsr' )  # should be 5: 3 r's, 2s's
print("result_score should be 5 ... and it is", result_score)
    
```

```python
#
# apply the scoring function to the strings in ALL_DATA
#

SCORED_ROWS = []

for row in ALL_DATA:
    # row[0] is the label ("human_or_machine" for the moment)
    label = row[0]
    # row[1] is the index (not used here, except in printing)
    index = row[1]
    # row[2] is the rps_string itself
    rps_string = row[2]
    # score it!
    result_score = score(rps_string)
    
    print(index, "    score:", result_score, rps_string[:42], "...")
    
    scored_row = [result_score, label, index, rps_string]
    SCORED_ROWS += [scored_row]
    
```

```python
#
# write out the scored rows to results.csv  (feel free to label by hand... or write another function!)
#

import csv

f = open("results.csv", "w", newline='')
w = csv.writer(f)
for row in SCORED_ROWS:
    w.writerow(row)
f.close()
```

```python
def myscore( rps_string ):
    """ associate a score with each rps-string 
        this scoring function simply counts the 9 different
        2 letter pairs
    """
    num_rr = count("rr", rps_string)
    num_rs = count("rs", rps_string)
    num_rp = count("rp", rps_string)
    num_pr = count("pr", rps_string)
    num_ps = count("ps", rps_string)
    num_pp = count("pp", rps_string)
    num_sr = count("sr", rps_string)
    num_ss = count("ss", rps_string)
    num_sp = count("sp", rps_string)
    result_list = [num_rr, num_rs, num_rp, num_pr, num_ps, num_pp, num_sr,num_ss, num_sp]
    expected_val = len(rps_string)/9
    list_of_errors = [abs(x-expected_val) for x in result_list]
    total_error = sum(list_of_errors)
    average_error = total_error/9
    return average_error

    

# let's test it out!
result_score = myscore( 'rpsrpsr' )  
print(result_score)
```
